```python
class Solution:
    def letterCombinations(self, digits: str):
        data = {
            2: ['a','b','c'],
            3: ['d','e','f'],
            4: ['g','h','i'],
            5: ['j','k','l'],
            6: ['m','n','o'],
            7: ['p','q','r','s'],
            8: ['t','u','v'],
            9: ['w','x','y','z']
        }
        
        if len(digits)  == 0:
            return []
        
        digits = [int(n) for n in digits]
        
        ans = ['']
        for n in digits:
            cur = []
            for w in ans:
                for new in data[n]:
                    cur.append(w+new)
            ans = cur[:]
        return ans

Solution().letterCombinations('234567')   
```
