```python
from Bio import SeqIO
from Bio.Seq import Seq

k = 3
seq_list = {}
answer = []

seq = SeqIO.parse('sample.fasta', 'fasta')
for s in seq:
    seq_list[s.name] = s.seq
    
seq_list
```

```python
for a, seq1 in seq_list.items():
    for b, seq2 in seq_list.items():
        if a != b and seq2.startswith(seq1[-k:]):
            print(f'{a} {b}')
            
```
