```python
arr = [a for a in input().split(' ')]
r = int(arr.pop())
arr.sort()
```

```python
def alpha(n):
    if n == 1:
        return arr
    return [nuc + a for nuc in alpha(n-1) for a in arr]

print('\n'.join(alpha(r)))
```

```python
from itertools import product
for p in product(arr, repeat=r):
    print(p)
```
