```python
from Bio import AlignIO
from Bio.motifs import Motif
from Bio import motifs
from Bio.Seq import Seq

alighment = AlignIO.read('HBA.aln', 'clustal')
instance = []
for record in alighment:
    s = Seq(str(record.seq))
    instance.append(s)

m = motifs.create(instance, alphabet= 'ARNDCEQGHILKMFPSTWYVX')
Motif.weblogo(m, 'HBA_WebLogo.png', alphabet = 'alphabet_protein')

from IPython.display import Image
Image('HBA_WebLogo.png')
```

```python
from Bio.Blast import NCBIWWW
from Bio import SeqIO

record = SeqIO.read('buccal_swab.unmapped1.fasta', format='fasta')
handle = NCBIWWW.qblast('blastn', 'nt', record.format('fasta'))

# result = handle.readlines()
# for s in result:
#     print(s)
```

```python
from Bio.Blast import NCBIXML

blast_record = NCBIXML.parse(handle)
E_VALUE = 0.05
for br in blast_record:
    for aln in br.alignments:
        for hsp in aln.hsps:
            if hsp.expect < E_VALUE:
                print(aln.title)
                print(aln.length)
                print(hsp.expect)
                print(hsp.query[0:75])
                print(hsp.match[0:75])
                print(hsp.sbjct[0:75])
```
